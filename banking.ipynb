```python
acc_pin_no="12345"
acc_balance=6000

pin=input("enter your pin no:")
if pin==acc_pin_no:
  print("welcome")
  print("\n 1.deposit \n 2.withdraw \n 3.checkbalance")
  choice=input("enter your choice")
  if choice=="1":
    amount=int(input("enter your amount"))
    acc_balance+=amount
    print("balance amount",acc_balance)
  elif choice=="2":
    amount=int(input("enter your amount"))
    if acc_balance>=amount:
      acc_balance-=amount
      print("the balance amount",acc_balance)
    else:
      print("insuffecient balance")
  else:
    print("balance amount",acc_balance)


else:
  print("pin incorrect")
```
